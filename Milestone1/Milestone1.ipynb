```python
def load_data(path):

    # Load dataset
    df = pd.read_csv(path, encoding="latin1")

    # Clean column names
    df.columns = df.columns.str.strip().str.lower()

    # Remove missing values
    df = df.dropna()

    # Remove duplicate rows
    df = df.drop_duplicates()

    # Select relevant columns
    df = df[['name',
             'category_list',
             'funding_total_usd',
             'funding_rounds',
             'country_code',
             'region',
             'status']]

    return df
```
